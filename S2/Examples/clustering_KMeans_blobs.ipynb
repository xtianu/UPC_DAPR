Generate data

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=100, centers=[(0,0),(2,2)],cluster_std=(1,1),n_features=2,random_state=0)

plt.scatter(X[:,0],X[:,1],c=y)
```

Apply k-means

```python
from sklearn.cluster import KMeans
NC = 2

km_algo = KMeans(NC).fit(X)
```

```python
km_algo.cluster_centers_ # final cluster centroids
```

```python
km_algo.labels_ #vector assigning observations to clusters
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=km_algo.labels_)
```

Now use any of the clustering algorithms in sklearn

https://scikit-learn.org/stable/modules/clustering.html#clustering


```python
from sklearn.cluster import AgglomerativeClustering

ag_algo = AgglomerativeClustering(n_clusters=2).fit(X)
ag_algo.labels_
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
Z = linkage(X)
dendrogram(Z)
```
